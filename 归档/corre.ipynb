```python
#读取 Tetuan City power consumption.csv
# DateTime,Temperature,Humidity,Wind Speed,general diffuse flows,diffuse flows,Zone 1 Power Consumption,Zone 2  Power Consumption,Zone 3  Power Consumption
# 1/1/2017 0:00,6.559,73.8,0.083,0.051,0.119,34055.6962,16128.87538,20240.96386
# 1/1/2017 0:10,6.414,74.5,0.083,0.07,0.085,29814.68354,19375.07599,20131.08434
# 1/1/2017 0:20,6.313,74.5,0.08,0.062,0.1,29128.10127,19006.68693,19668.43373


data = []
with open('Tetuan City power consumption.csv') as f:
    for line in f:
        data.append(line.strip().split(','))
    
for i in range(1,len(data)):
    for j in range(1,len(data[i])):
        data[i][j] = float(data[i][j])

print(data[1])

['DateTime', 'Temperature', 'Humidity', 'Wind Speed', 'general diffuse flows', 'diffuse flows', 'Zone 1 Power Consumption', 'Zone 2  Power Consumption', 'Zone 3  Power Consumption']

['1/1/2017 0:00', 6.559, 73.8, 0.083, 0.051, 0.119, 34055.6962, 16128.87538, 20240.96386]



import matplotlib.pyplot as plt
import datetime
# 计算平均值用于归一化
Temperature = 0
Humidity = 0
WindSpeed = 0
generaldiffuseflows = 0
diffuseflows = 0
Zone1  = 0
Zone2 = 0
Zone3 = 0

for i in range(1, len(data)):
    Temperature += data[i][1]
    Humidity += data[i][2]
    WindSpeed += data[i][3]
    generaldiffuseflows += data[i][4]
    diffuseflows += data[i][5]
    Zone1 += data[i][6]
    Zone2 += data[i][7]
    Zone3 += data[i][8]

Temperature /= len(data)
Humidity /= len(data)
WindSpeed /= len(data)
generaldiffuseflows /= len(data)
diffuseflows /= len(data)
Zone1 /= len(data)
Zone2 /= len(data)
Zone3 /= len(data)

for i in range(1, len(data)):
    # '1/1/2017 0:00'
    Fdate = data[i][0]
    Fdate = Fdate.split(' ')
    h = Fdate[1].split(':')[0]
    m = Fdate[1].split(':')[1]
    data[i].append((int(h)*60+int(m))/10)
    Fdate = Fdate[0].split('/')
    #根据日期计算星期几
    date_object = datetime.datetime(int(Fdate[2]), int(Fdate[0]), int(Fdate[1]))
    weekday = date_object.weekday()
    data[i].append(weekday)
    del data[i][0]
    

for i in range(1, 10):
    print(data[i])

print(Temperature, Humidity, WindSpeed, generaldiffuseflows, diffuseflows, Zone1, Zone2, Zone3)

```

```python
def pearson_correlation_coefficient(x, y):
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x_squared = sum(xi**2 for xi in x)
    sum_y_squared = sum(yi**2 for yi in y)
    sum_xy = sum(x[i] * y[i] for i in range(n))
    numerator = n * sum_xy - sum_x * sum_y
    denominator = ((n * sum_x_squared - sum_x**2) * (n * sum_y_squared - sum_y**2))**0.5
    if denominator == 0:
        return 0  # Avoid division by zero
    r = numerator / denominator
    return r

te = []
hu = []
wi = []
ge = []
di = []
zo1 = []
zo2 = []
zo3 = []
ti = []
we = []


# 创建10乘10的矩阵，用于存储皮尔逊相关系数
out = [[0 for i in range(10)] for j in range(10)]


for i in range(1, len(data)):
    te.append(data[i][0])
    hu.append(-1* (data[i][1] - Humidity))
    wi.append(data[i][2])
    ge.append(data[i][3])
    di.append(data[i][4])
    zo1.append(data[i][5])
    zo2.append(data[i][6])
    zo3.append(data[i][7])
    ti.append(data[i][8])
    we.append((data[i][9] - 3)* -1)



###################################
# # 将0-143时间信息分段表示
# section = 12
# for i in range(1, len(data)):
#     ti[i-1] = int(ti[i-1]/section)
# print(ti)
# # 根据时间序列重新排序所有数据

# # 使用zip将列表组合起来，然后根据第一个列表(list_a)的值排序
# # 这里使用lambda函数作为sorted的key参数，指定按照元组中第一个元素排序
# combined_lists = sorted(zip(ti,te, hu, wi, ge, di, zo1, zo2, zo3), key=lambda x: x[0])

# # 解压缩排序后的列表，以恢复到原始的分开的列表格式
# ti, te, hu, wi, ge, di, zo1, zo2, zo3 = zip(*combined_lists)

# # 将结果从元组转换回列表
# ti = list(ti)
# te = list(te)
# hu = list(hu)
# wi = list(wi)
# ge = list(ge)
# di = list(di)
# zo1 = list(zo1)
# zo2 = list(zo2)
# zo3 = list(zo3)
##################################


# 为了方便，创建一个列表，包含所有x变量
x_variables = [te, hu, wi, ge, di, ti, we,zo1, zo2, zo3]

# 计算y与每个x变量的皮尔逊相关系数
for i, x in enumerate(x_variables, start=1):
    for j, y in enumerate(x_variables, start=1):
        r = pearson_correlation_coefficient(x, y)
        out[i-1][j-1] = r

print(out)



import numpy as np
import matplotlib.pyplot as plt

# 将主对角线设置为NaN
out = np.array(out)
np.fill_diagonal(out, np.nan)

# 获取矩阵的最大绝对值
vmax = np.nanmax(np.abs(out))

maxtime = 143






# 绘制热力图
plt.matshow(out, cmap='coolwarm', vmin=-vmax, vmax=vmax)
plt.colorbar()

# 添加x轴和y轴的标签
pltLb = ['Temperature', 'Humidity', 'Wind Speed', 'general diffuse flows', 'diffuse flows','Time','Weekday', 'Zone 1', 'Zone 2', 'Zone 3']
plt.xticks(range(len(pltLb)), pltLb, rotation=90)
plt.yticks(range(len(pltLb)), pltLb)

plt.show()
```
